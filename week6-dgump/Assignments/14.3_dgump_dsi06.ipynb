```python
def missingness_summary(df, print_log=False, sort=None):
    missings = df.isna().mean()

    if sort == "asc":
        missings = missings.sort_values()
    elif sort == "desc":
        missings = missings.sort_values(ascending=False)

    if print_log:
        print(missings)

    return missings


us_edu_missing_report = missingness_summary(us_edu, print_log=True, sort="desc")
```

```python
edu_missing_counts = us_edu.isna().sum(axis=1)
len(edu_missing_counts)
```

```python
edu_dropped = us_edu
dropped_indices = []
for index in us_edu.index:
    if edu_missing_counts[index] > 12:
        # print(index)
        dropped_indices.append(index)
        edu_dropped = edu_dropped.drop(labels=index, axis=0)
edu_dropped.info()
len(dropped_indices)
```
