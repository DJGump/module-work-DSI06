## For two of the following distributions:

* Normal

* Binomial

* Gamma

* Poisson

### 1. Create a line chart plotting the sample mean as more and more trials are conducted, up to 1,000 trials.

### 2. Randomly sample the mean 1,000 times and plot the resulting frequency of sample means as a histogram.

#### Distribution 1: Binomial

   1. Create a line chart plotting the sample mean as more and more trials are conducted, up to 1,000 trials.

```python
# binomial distribution with 5 events and prob = 0.3
means = []
# must start range at 1 as you can't have 0 trials
for trials in range(1,1000):
    exp = np.random.binomial(5,0.3,trials)
    exp_mean = exp.mean()
    means.append(exp_mean)
large_num = pd.DataFrame({'num_trials' : means})

large_num.plot()
plt.xlabel('trials')
plt.ylabel('sample mean')
plt.copper()
plt.show()
```

   2. Randomly sample the mean 1,000 times and plot the resulting frequency of sample means as a histogram.

```python
# generate 1,000 samples, store the mean, 1,000 times
means = [mean(np.random.binomial(5, 0.3, 1000)) for _ in range(1000)]
plt.hist(means)
plt.xlabel('mean')
plt.ylabel('frequency')
plt.show()
```

### Distribution 2: Gamma

#### 1. Create a line chart plotting the sample mean as more and more trials are conducted, up to 1,000 trials.

```python
# gamma distribution with shape 1 and scale 1
means = []

for trials in range(1,1000):
# there appears to be a 'mean = shape * scale' relationship with gamma
    exp = np.random.gamma(2,100,trials)
    exp_mean = exp.mean()
    means.append(exp_mean)
large_num = pd.DataFrame({'num_trials' : means})

large_num.plot()
plt.xlabel('trials')
plt.ylabel('sample mean')
plt.show()
```

#### 2. Randomly sample the mean 1,000 times and plot the resulting frequency of sample means as a histogram.

```python
# sample the mean of 1000 trials, 1000 times
means = [mean(np.random.gamma(2, 100, 1000)) for _ in range(1000)]


plt.hist(means)
plt.xlabel('mean')
plt.ylabel('frequency')
plt.show()
```
